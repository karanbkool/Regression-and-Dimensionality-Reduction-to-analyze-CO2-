# Dataset Manipulation

```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pylab as pl
```

```python
pd.set_option("max_columns", 34)
pd.set_option("max_rows", 100)
df = pd.read_csv('data analytics-car emission.csv',encoding='windows-1254')
df.head()
```

# Cleaning Dataset

```python
#removing last 7 columns
df = df.drop(['Unnamed: 27','Unnamed: 28','Unnamed: 29','Unnamed: 30','Unnamed: 31','Unnamed: 32','Unnamed: 33'], axis=1)
df.head(20)
```

```python
#removing columns with nan values
df = df.drop(['Electric energy consumption Miles/kWh','wh/km','Maximum range (Km)','Maximum range (Miles)','Electricity cost','THC Emissions [mg/km]'], axis=1)
df.head(10)
```

```python
#removing redundant and unnecessary columns (Description, FuelCost, Imperial Urban, Imperial Extra-Urban)
df = df.drop(['Description', 'Metric Urban (Cold)', 'Metric Extra-Urban', 'Imperial Urban (Cold)', 'Imperial Extra-Urban', 'Fuel Cost 12000 Miles'], axis=1)
df.head(10)
```

```python
no_data=df.isnull()
for column in no_data.columns.values.tolist():
    print(column)
    print(no_data[column].value_counts())
    print('----------------------------------------------------')
```

```python
#particulates and THC have too many missing values so we remove them. The rest we replace nan with the mean of the column
df = df.drop(['Particulates [No.] [mg/km]','THC + NOx Emissions [mg/km]'], axis=1)
df.head()
```

```python
#checking mode of categorical data
print(df['Transmission'].value_counts())
print('------------------------------')
df['Fuel Type'].value_counts()
```

# Replacing Nan Values with Mean and Mode

```python
#assigning mean and mode vals
mode_ft ="Diesel"
mode_t = "M6"
mean_NO = df['Emissions NOx [mg/km]'].mean()
mean_CO = df['Emissions CO [mg/km]'].mean()
mean_ec = df['Engine Capacity'].mean()
mean_mc = df['Metric Combined'].mean()
mean_ic = df['Imperial Combined'].mean()
df['Total cost / 12000 miles'] = df['Total cost / 12000 miles'].replace('[\£,]', '', regex=True).astype(float)
mean_tct = df['Total cost / 12000 miles'].mean()
#replacing nan values with mean and mode
df["Fuel Type"].replace(np.nan, mode_ft, inplace=True)
df["Transmission"].replace(np.nan, mode_t, inplace=True)
df["Emissions NOx [mg/km]"].replace(np.nan, mean_NO, inplace=True)
df["Emissions CO [mg/km]"].replace(np.nan, mean_CO, inplace=True)
df["Engine Capacity"].replace(np.nan, mean_ec, inplace=True)
df["Metric Combined"].replace(np.nan, mean_mc, inplace=True)
df["Imperial Combined"].replace(np.nan, mean_ic, inplace=True)
df["Total cost / 12000 miles"].replace(np.nan, mean_tct, inplace=True)
```

Here we can see that the count of all columns is equal

# Scaling and Data Preparation

```python
import statsmodels
from statsmodels.stats.outliers_influence import variance_inflation_factor
def calc_vif(x):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["variables"] = x.columns
    vif["VIF"] = [variance_inflation_factor(x.values, i) for i in range(x.shape[1])]

    return(vif)
```

```python
#Engine Capacity 	Metric Combined 	Imperial Combined 	CO2 g/km 	Euro Standard 	Noise Level dB(A) 	Emissions CO [mg/km] 	Emissions NOx [mg/km]
x = df[['Engine Capacity','Metric Combined','Imperial Combined','Noise Level dB(A)','Emissions CO [mg/km]','Emissions NOx [mg/km]','Total cost / 12000 miles']]
y = df[['CO2 g/km']]

calc_vif(x)
```

We can see here that Variance Inflation Factor (VIF) is greater than 5 for most variables. aIdeally it should be lesser than 5. This is a sign of <b>multicolinearity</b>.

# Estimating and Removing Multicolinearity

```python
import sklearn
from sklearn.preprocessing import StandardScaler

x = StandardScaler().fit_transform(x)
y = StandardScaler().fit_transform(y)
```

```python
df2 = pd.DataFrame(x, columns=['Engine Capacity','Metric Combined','Imperial Combined','Noise Level dB(A)','Emissions CO [mg/km]','Emissions NOx [mg/km]','Total cost / 12000 miles'])
calc_vif(df2)
```

Here we see that after standardization of data, the VIF has been reduced. This is done by removing structural multicollinearity.

```python
# removing metric combined from the features
x = df2[['Engine Capacity','Imperial Combined','Noise Level dB(A)','Emissions CO [mg/km]','Emissions NOx [mg/km]','Total cost / 12000 miles']] 
calc_vif(x)
```

# Creating Testing and Training Datasets

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=38)
```

#  Multiple Linear Regression

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
# predicting the test dataset 
yhat_m = lr.predict(x_test)
print('Intercept: ', lr.intercept_)
print('Coefficients: ', lr.coef_)
```

<b> Multiple Regression Equation</b> 
<br><br>
CO2 Emissions (g/km) = 0.00509832 + 0.04371917 * (Engine Capacity) - 0.12302398 * (Imperial Combined) - 0.00567771 * (Noise Level dB(A)) - 0.02891212  * (Emissions CO) + 0.11177254 * (Emissions NOx) + 0.87919599 * (Total Cost/12000 km)

```python
#data visualization using distribution plot
def DistributionPlot(Red, Blue, RedName, BlueName, Title):
    plt.figure(figsize=(9,6))

    ax1 = sns.distplot(Red, hist=False, color="crimson", label=RedName)
    ax2 = sns.distplot(Blue, hist=False, color="lightseagreen", label=BlueName, ax=ax1)

    plt.title(Title)

    plt.show()
    plt.close()
```

```python
DistributionPlot(y_test, yhat_m, 'Actual Values', 'Predicted Values', 'Distribution Plot of Multiple Linear Regression')
```

```python
n = len(x_test)-1
p = 6
```

```python
# model evaluation
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
print("MSE: %.5f" % mean_squared_error(y_test,yhat_m))
print("RMSE: %.5f" % np.sqrt(mean_squared_error(y_test,yhat_m)))
print("R-squared value: %.5f" % r2_score(y_test,yhat_m))
print("Adjusted R-Squared Value", round(1-(1-r2_score(y_test, yhat_m))*((n)/(n- p )), 5))
```

# Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
# using function to find the best degree of polynomial
min_d = 0
min_mse= 200000000000
mses = []
degrees = np.arange(1,10)
for deg in degrees:

    # Train features
    polyreg = PolynomialFeatures(degree=deg, include_bias=False)
    px_train = polyreg.fit_transform(x_train)

    # Linear regression
    lr = LinearRegression()
    lr.fit(px_train, y_train)

    # Compare with test data
    px_test = polyreg.fit_transform(x_test)
    yhat_p = lr.predict(px_test)
    p_mse = mean_squared_error(y_test, yhat_p)
    mses.append(p_mse)
    
    # Cross-validation of degree
    if min_mse > p_mse:
        min_mse = p_mse
        min_deg = deg

# Plot and present results
print('Best degree is {} with MSE {}'.format(min_deg, min_mse))
        
fig = plt.figure(figsize=(9,6))
ax = fig.add_subplot(111)
ax.plot(degrees, mses)
ax.set_yscale('log')
ax.set_xlabel('Degree')
ax.set_ylabel('MSE')
```

Polynomial Regression with degree 1 is Simple Linear Regression. The lowest Mean square error is for degree=3. We will proceed with that 

```python
# Multivariate Polynomial Regression of degree 3
polyreg = PolynomialFeatures(degree=3, include_bias=False)
px_train = polyreg.fit_transform(x_train)
    
lr = LinearRegression()    
lr.fit(px_train, y_train)

px_test = polyreg.fit_transform(x_test)
yhat_p = lr.predict(px_test)
```

```python
print('Intercept: ', lr.intercept_)
print('Coefficients: ', lr.coef_)
```

```python
DistributionPlot(y_test, yhat_p, 'Actual Values', 'Predicted Values', 'Distribution Plot of Polynomial Linear Regression with degree 3')
```

This model is overfitting. So, we use degree= 2.

# Using PCA for Dimensionality Reduction

```python
#df_e is the feature selection
x2 = df[['Engine Capacity','Imperial Combined','Noise Level dB(A)','Emissions CO [mg/km]','Emissions NOx [mg/km]','Total cost / 12000 miles']]
x2 = StandardScaler().fit_transform(x2)
```

```python
# we perform dimensionality reduction to reduce number of predictors to 6
from sklearn.decomposition import PCA
pca = PCA(n_components = 5)
pc_dt = pca.fit_transform(x2)

x2 = pd.DataFrame(data=pc_dt, columns=['PC1','PC2','PC3','PC4','PC5'])
```

```python
x2.head()
```

```python
calc_vif(x2)
```

```python
print(pca.explained_variance_ratio_)
print("Cumulative explanation with each variable:",pca.explained_variance_ratio_.cumsum())
```

# Using Regression after PCA feature selection

```python
x_train2, x_test2, y_train2, y_test = train_test_split(x2, y, test_size=0.3, random_state=38)
```

```python
lr = LinearRegression()
lr.fit(x_train2,y_train2)
```

```python
print(lr.intercept_)
print(lr.coef_)
```

<b> Multiple Regression Equation with Dimensionality Regression</b> 
<br><br>
CO2 Emissions (g/km) = 0.00827923 - 0.50397578 * (PC1) + 0.15372517 * (PC2) - 0.10301163 * (PC3) + 0.0471757 * (PC4) - 0.00652214  * (PC5) + 0.39536759 * (PC6)

```python
yhat_pcr = lr.predict(x_test2)
```

```python
DistributionPlot(y_test, yhat_pcr, 'Actual Values', 'Predicted Values', 'Distribution Plot of MLR with Dimensionality Reduction')
```

The predicted model is overfitting so the r-squared value can be misleading.

```python
print("MSE: %.5f" % mean_squared_error(y_test, yhat_pcr))
print("RMSE: %.5f" % np.sqrt(mean_squared_error(y_test, yhat_pcr)))
print("R-squared value: %.5f" % r2_score(y_test, yhat_pcr))
print("Adjusted R-Squared Value", round(1-(1-r2_score(y_test, yhat_pcr))*((n)/(n - p)), 5))
```
